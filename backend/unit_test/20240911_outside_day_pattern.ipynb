```python
import pandas as pd
```

```python
from backend.data_center.kline_data.kline_data_collector import *
tv = KlineDataCollector()
file_abspath = tv.get_abspath(symbol='BTC', interval=Interval.in_daily)
file_abspath
```

```python
df = pd.read_csv(f"{file_abspath}")
df
```

```python
df['prev_high'] = df['high'].shift(1)  # 前一天的 high
df['prev_low'] = df['low'].shift(1)    # 前一天的 low
df['prev2_high'] = df['high'].shift(2) # 前两天的 high
df['prev2_low'] = df['low'].shift(2)   # 前两天的 low
df
```

```python
df['is_outside'] = (df['prev_high'] > df['prev2_high']) & (df['close'].shift(1) < df['prev2_low'])
```

```python
df.drop(['prev_high', 'prev_low', 'prev2_high', 'prev2_low'], axis=1, inplace=True)
```

```python
# 前一天出现了外移日的行情。多少为增加？
df[df['is_outside'] == True]
```

```python
df['is_increase'] = df['close'] > df['open']
```

```python
df[df['is_outside'] == True]['is_increase'].sum()
```
