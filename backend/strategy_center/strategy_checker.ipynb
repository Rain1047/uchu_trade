```python
from backend.strategy_center.atom_strategy.exit_strategy.dbb_exit_strategy import dbb_exit_strategy_for_backtest

tv = KlineDataCollector()
file_abspath = tv.get_abspath(symbol='BTC', interval=Interval.in_daily)
df = pd.read_csv(f"{file_abspath}")

# 确保datetime格式正确
df['datetime'] = pd.to_datetime(df['datetime'])
df = dbb_entry_long_strategy_backtest(df)
df = dbb_exit_strategy_for_backtest(df)
df = df[df['datetime'] > "2023-12-31 08:00:00"]
df
```

```python



# 选择这几个时间点及其前后的数据点以便检查
selected_times = pd.to_datetime([
    "2024-01-04 08:00:00",
    "2024-01-03 08:00:00",
    "2024-01-02 08:00:00"
])

# 获取这些时间点及其前两个时间点的数据
all_needed_data = df[['datetime', 'open', 'close', 'upper_band1']].copy()
print("\n完整的条件检查：")
for dt in selected_times:
    current_row = all_needed_data[all_needed_data['datetime'] == dt]
    if len(current_row) > 0:
        idx = current_row.index[0]
        print(f"\n当前时间: {dt}")
        print(f"当前行:")
        print(f"open < upper_band1: {current_row['open'].iloc[0]} < {current_row['upper_band1'].iloc[0]} = {current_row['open'].iloc[0] < current_row['upper_band1'].iloc[0]}")
        print(f"close > upper_band1: {current_row['close'].iloc[0]} > {current_row['upper_band1'].iloc[0]} = {current_row['close'].iloc[0] > current_row['upper_band1'].iloc[0]}")
        
        # 获取前一行数据
        prev_row1 = all_needed_data.shift(1).loc[idx]
        print(f"\n前一行:")
        print(f"prev_open < prev_upper_band1: {prev_row1['open']} < {prev_row1['upper_band1']} = {prev_row1['open'] < prev_row1['upper_band1']}")
        
        # 获取前两行数据
        prev_row2 = all_needed_data.shift(2).loc[idx]
        print(f"\n前两行:")
        print(f"prev_open_2 < prev_upper_band1_2: {prev_row2['open']} < {prev_row2['upper_band1']} = {prev_row2['open'] < prev_row2['upper_band1']}")
        
        print("\n最终entry_sig值:", df.loc[idx, 'entry_sig'])
        print("-" * 80)
        
df[df['entry_sig'] == 1]
# df[df['sell_sig'] == 1]
```
